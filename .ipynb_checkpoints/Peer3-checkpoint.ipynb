## The "Quantum Equlibrium-Disequlibrium" (QED) IRL-based model of stock returns

In Week 3 lectures of our course we presented the "QED" model
$$
d X_t = \kappa X_t  \left( \frac{\theta}{\kappa} - X_t - \frac{g}{\kappa} X_t^2 \right) dt +  X_t \left( {\bf w} {\bf z}_t \, dt + \sigma d W_t \right)
$$

In this project, you will explore calibration of this model to market data.
As in the course project for course 3 you analyzed the same model in the limit $ g = 0 $, you would be able to re-utilize parts of your previous code in this project).

## Part 1: Model calibration with or without moving average signals (Max 10 points)

To calibrate the model, it is convenient to use the log-prices instead of prices. Diffusion in the log-space 
$ y = \log x $ is given by the following Langevin equation:

$$
d y_t = - \frac{ \partial V(y)}{\partial y} dt + \sigma dW_t , \; \; \;  V(y)  \equiv - \left( \theta - \frac{\sigma^2}{2} + {\bf w} {\bf z}_t \right) y  + \kappa e^y + \frac{1}{2} g e^{2y}
$$

where $ W_t $ is a standard Brownian motion.
In terms of variables $ y = \log x $, the negative log-likelihood of data is therefore

$$
LL_M (\Theta) = - \log \prod_{t=0}^{T-1} 
\frac{1}{ \sqrt{ 2 \pi  \sigma^2}  } 
\exp \left\{  - \frac{1}{2 \sigma^2} \left(   \frac{ y_{t+ \Delta t} -   y_{t}}{ \Delta t} +  \frac{ \partial V(y)}{\partial y}   
\right)^2
\right\} , 
$$ 

where $ {\bf y}_t  = \log x_t $  now stands for observed values of log-cap. Note that because the model is Markov, the product over $ t = 0, \ldots, T-1 $ does not 
necessarily mean a product of transitions along the same trajectory. The negative log-likelihood should be minimized to estimate parameters $ 
\theta $, $ \sigma $, $ \kappa $, $ g $ and  $ {\bf w} $. You can try to estimate the model first without signals, then with signals.


```python
# Put the rest of you code and analysis for Part I here 
```

$$
LL_M (\Theta) = - \log \prod_{t=0}^{T-1} 
\frac{1}{ \sqrt{ 2 \pi  \sigma^2}  } 
\exp \left\{  - \frac{1}{2 \sigma^2} \left(   \frac{ y_{t+ \Delta t} -   y_{t}}{ \Delta t} +  \frac{ \partial V(y)}{\partial y}   
\right)^2
\right\} 
$$ 

$$ = \frac{1}{2 \sigma^2} \sum_{t=0}^{T-1}  \left( \frac{ y_{t+ \Delta t} - y_{t}}{ \Delta t} +  \frac{ \partial V(y)}{\partial y} \right)^2  + \frac{T}{2} \log (2 \pi \sigma^2) $$

$$ = \frac{1}{2 \sigma^2} \sum_{t=0}^{T-1}  \left( \frac{ y_{t+ \Delta t} - y_{t}}{ \Delta t} + \left( \theta - \frac{\sigma^2}{2} + {\bf w} {\bf z}_t \right) + \kappa e^y + g e^{2y} \right)^2  + \frac{T}{2} \log (2 \pi \sigma^2) $$

```python
y = np.log(df_cap)
z1 = y.rolling(window=10).mean()
z2 = y.rolling(window=30).mean()

delta_y = (y.shift(-1) - y)

def extract_params(params):
    theta = params[0]
    sigma = np.abs(params[1]) # Enforce sigma positive
    kappa = 1e-16 * params[2]
    g = 1e-32 * params[3]
    w1 = params[4]
    w2 = params[5]

    return theta, sigma, kappa, g, w1, w2


def neg_log_likelihood(stock_name):
    def f(params):
        theta, sigma, kappa, g, w1, w2 = extract_params(params)

        k = delta_y[stock_name][30:-1] + \
            -(theta + (sigma**2 / 2) + w1 * z1[stock_name][30:-1] + w2 * z2[stock_name][30:-1]) + \
            kappa * np.exp(y[stock_name][30:-1]) + \
            g * np.exp(2 * y[stock_name][30:-1])
        
        return np.sum(1 / (sigma ** 2) * np.sum(k**2, axis=0) + (y.shape[0] - 31) / 2 * np.log(2 * np.pi * sigma ** 2))
        
    return f

from scipy.optimize import minimize

k = y.shape[1]
stocks = y.columns

result = pd.DataFrame(columns=['stock', 'theta', 'sigma', 'kappa', 'g', 'w1', 'w2'])

for stock_num in range(k):
    stock_name = stocks[stock_num]
    print(stock_name)

    x0 = np.ones(6)
    res = minimize(
        neg_log_likelihood(stock_name), x0, method='nelder-mead', 
        options={
            'xtol': 1e-8,
            'disp': True,
            'maxiter': 25000
        }
    )
    
    theta, sigma, kappa, g, w1, w2 = extract_params(res.x)
    
    result = result.append({
        'stock': stock_name, 
        'theta': theta, 
        'sigma': sigma, 
        'kappa': kappa, 
        'g': g,
        'w1': w1,
        'w2': w2
    }, ignore_index=True)
    
result = result.set_index('stock')
```

```python
# Values obtained for the fit for each stock:
result
```

```python
# Next, we compute the predicted values for each stock and day:
y_predict = pd.DataFrame().reindex_like(y)

for stock_num in range(k):
    stock_name = stocks[stock_num]

    theta, sigma, kappa, g, w1, w2 = result.loc[stock_name]
    
    y_predict[stock_name] = y[stock_name] + \
        -(theta - sigma**2 / 2 + w1 * z1[stock_name] + w2 * z2[stock_name]) + \
        kappa * np.exp(y[stock_name]) + \
        (1/2) * g * np.exp(2 * y[stock_name])
        
y_predict = y_predict.shift(1)

ticker = 'AAPL'
start_date = '2010-02-15'
end_date = '2017-12-21'

df_predict = np.exp(y_predict)
df_baseline = df_cap.shift(1)

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)

ax.plot(
    df_cap.loc[start_date:end_date, :].index, 
    df_predict.loc[start_date:end_date, 'AAPL'] / df_cap.loc[start_date:end_date, 'AAPL'], 
    label='Predicted cap / cap'
)

ax.legend(loc='best')
ax.set_ylabel('AAPL Prediction cap / cap')

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)

ax.plot(
    df_cap.loc[start_date:end_date, :].index, 
    df_predict.loc[start_date:end_date, 'AAPL'] / df_baseline.loc[start_date:end_date, 'AAPL'], 
    label='Predicted cap / previous cap'
)

ax.legend(loc='best')
ax.set_ylabel('AAPL Previous cap / cap')

plt.show()
```

## Part 2: Analysis of default rates  (Max 10 point)

For a particle in a potential $ V(y) $ with a metastable minimum $ y = a $ and a barrier with a peak located at $ y = b $, the famous Kramers' escape formula gives the following expression for the escape rate $ r $ (see e.g. the book by van Kampen):

$$ 
r = \frac{\sqrt{ V''(a) \left| V''(b) \right| }}{2 \pi} \exp \left[ - \frac{2}{\sigma^2} (V(b) - V(a) ) \right]
$$

Here $ V''(a) $ and $ V''(b) $ stand for the second derivatives of the potential $ V(y) $ at the minimum and the maximum, respectively. This formula applies as long as the barrier height $ \Delta E \equiv  (V(b) - V(a) \gg \frac{\sigma^2}{2} $. 

Apply the Kramers formula to the QED potential and parameters that you found in your calibration. What range of values of $ r $ do you obtain? Do these values make sense to you? Can you think how you could use the Kramers relation as a way to regularize your MLE calibration?

The maximum and minimum of $ V(y) $ are obtained at $ \frac{ \partial V(y)}{\partial y} = 0 $:

$$
y = \log \left( \frac{-\kappa \pm \sqrt{\kappa^2 + 4 g \alpha}}{2g} \right)
$$

where

$$
V(y) = - \alpha y  + \kappa e^y + \frac{1}{2} g e^{2y}
$$

$$
V''(y) = \kappa e^y + 2 g e^{2y}
$$

$$
\alpha = \theta - \frac{\sigma^2}{2} + {\bf w} {\bf z}_t
$$

```python
# We will redo the fits without the signals:
delta_y = (y.shift(-1) - y)

def extract_params_no_signals(params):
    theta = params[0]
    sigma = np.abs(params[1]) # Enforce sigma positive
    kappa = 1e-16 * params[2]
    g = 1e-32 * params[3]

    return theta, sigma, kappa, g


def neg_log_likelihood_no_signals(stock_name):
    def f(params):
        theta, sigma, kappa, g = extract_params_no_signals(params)

        k = delta_y[stock_name][30:-1] + \
            -(theta + (sigma**2 / 2)) + \
            kappa * np.exp(y[stock_name][30:-1]) + \
            g * np.exp(2 * y[stock_name][30:-1])
        
        return np.sum(1 / (sigma ** 2) * np.sum(k**2, axis=0) + (y.shape[0] - 31) / 2 * np.log(2 * np.pi * sigma ** 2))
        
    return f

from scipy.optimize import minimize

k = y.shape[1]
stocks = y.columns

result_no_signals = pd.DataFrame(columns=['stock', 'theta', 'sigma', 'kappa', 'g'])

for stock_num in range(k):
    stock_name = stocks[stock_num]
    print(stock_name)

    x0 = np.ones(4)
    res = minimize(
        neg_log_likelihood_no_signals(stock_name), x0, method='nelder-mead', 
        options={
            'xtol': 1e-8,
            'disp': True,
            'maxiter': 25000
        }
    )
    
    theta, sigma, kappa, g = extract_params_no_signals(res.x)
    
    result_no_signals = result_no_signals.append({
        'stock': stock_name, 
        'theta': theta, 
        'sigma': sigma, 
        'kappa': kappa, 
        'g': g
    }, ignore_index=True)
    
result_no_signals = result_no_signals.set_index('stock')
```

```python
result_no_signals
```

```python
# We can now compute  𝑎 ,  𝑏 , and  𝑟  for each of these, as well as the barrier height  Δ𝐸 , where defined:
df = result_no_signals.copy(deep=True)
df['a'] = np.zeros(k)
df['b'] = np.zeros(k)
df['r'] = np.zeros(k)
df['delta E'] = np.zeros(k)
df["sigma squared / 2"] = np.zeros(k)

for stock_num in range(k):
    stock_name = stocks[stock_num]
    theta, sigma, kappa, g = result_no_signals.loc[stock_name]
    alpha = theta - sigma**2 / 2
    
    y_plus = np.log((np.sqrt(kappa**2 + 4 * g * alpha) - kappa) / (2 * g))
    y_minus = np.log((-np.sqrt(kappa**2 + 4 * g * alpha) - kappa) / (2 * g))
    a = min(y_plus, y_minus)
    b = max(y_plus, y_minus)
    
    V_a = -alpha * a + kappa * np.exp(a) + (1/2) * g * np.exp(2 * a)
    V_b = -alpha * b + kappa * np.exp(b) + (1/2) * g * np.exp(2 * b)
    
    V_pp_a = kappa * np.exp(a) + 2 * g * np.exp(2 * a)
    V_pp_b = kappa * np.exp(b) + 2 * g * np.exp(2 * b)
    
    r = np.sqrt(V_pp_a * np.abs(V_pp_b)) / (2 * np.pi) * np.exp((-2 / (sigma**2)) * (V_b - V_a))
    
    delta_e = V_b - V_a
    
    df['a'][stock_name] = a
    df['b'][stock_name] = b
    df['r'][stock_name] = r
    df['delta E'][stock_name] = delta_e
    df['sigma squared / 2'][stock_name] = sigma**2 / 2
    
df[df['delta E'] > df['sigma squared / 2']]
```

```python
# Put the rest of your code and analysis for Part 3 here.
# Let's add a signal that represents the average log market cap weight:
z_mkt = y.mean(axis=1)
delta_y = (y.shift(-1) - y)

def extract_params_part_3(params):
    theta = params[0]
    sigma = np.abs(params[1]) # Enforce sigma positive
    kappa = 1e-16 * params[2]
    g = 1e-32 * params[3]
    w_mkt = params[4]
    
    return theta, sigma, kappa, g, w_mkt


def neg_log_likelihood_part_3(stock_name):
    def f(params):
        theta, sigma, kappa, g, w_mkt = extract_params_part_3(params)

        k = delta_y[stock_name][:-1] + \
            -(theta + (sigma**2 / 2) + w_mkt * z_mkt[:-1]) + \
            kappa * np.exp(y[stock_name][:-1]) + \
            g * np.exp(2 * y[stock_name][:-1])
            
        
        return np.sum(1 / (sigma ** 2) * np.sum(k**2, axis=0) + (y.shape[0] - 1) / 2 * np.log(2 * np.pi * sigma ** 2))
        
    return f

from scipy.optimize import minimize

k = y.shape[1]
stocks = y.columns

columns = ['stock', 'theta', 'sigma', 'kappa', 'g', 'w_mkt']    
result_part_3 = pd.DataFrame(columns=columns)

for stock_num in range(k):
    stock_name = stocks[stock_num]
    print(stock_name)

    x0 = np.zeros(8)
    # Load sensible starting point from no-signals exercise
    x0[0] = result_no_signals['theta'][stock_name]
    x0[1] = result_no_signals['sigma'][stock_name]
    x0[2] = result_no_signals['kappa'][stock_name]
    x0[3] = result_no_signals['g'][stock_name]
    
    res = minimize(
        neg_log_likelihood_part_3(stock_name), x0, method='nelder-mead', 
        options={
            'xtol': 1e-8,
            'disp': True,
            'maxiter': 25000
        }
    )
    
    theta, sigma, kappa, g, w_mkt = extract_params_part_3(res.x)
    result_part_3 = result_part_3.append({
        'stock': stock_name, 
        'theta': theta, 
        'sigma': sigma, 
        'kappa': kappa, 
        'g': g,
        'w_mkt': w_mkt
    }, ignore_index=True)
    
result_part_3 = result_part_3.set_index('stock')
```

```python
y_predict = pd.DataFrame().reindex_like(y)

for stock_num in range(k):
    stock_name = stocks[stock_num]

    theta, sigma, kappa, g, w_mkt = result_part_3.loc[stock_name]
        
    y_predict[stock_name] = y[stock_name] + \
        -(theta - sigma**2 / 2 + w_mkt * z_mkt) + \
        kappa * np.exp(y[stock_name]) + \
        (1/2) * g * np.exp(2 * y[stock_name])
        
        
y_predict = y_predict.shift(1)

ticker = 'AAPL'
start_date = '2010-02-15'
end_date = '2017-12-21'

df_predict = np.exp(y_predict)
df_baseline = df_cap.shift(1)

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)

ax.plot(
    df_cap.loc[start_date:end_date, :].index, 
    df_predict.loc[start_date:end_date, 'AAPL'] / df_cap.loc[start_date:end_date, 'AAPL'], 
    label='Predicted cap / cap'
)

ax.legend(loc='best')
ax.set_ylabel('AAPL Prediction cap / cap')

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)

ax.plot(
    df_cap.loc[start_date:end_date, :].index, 
    df_predict.loc[start_date:end_date, 'AAPL'] / df_baseline.loc[start_date:end_date, 'AAPL'], 
    label='Predicted cap / previous cap'
)

ax.legend(loc='best')
ax.set_ylabel('AAPL Previous cap / cap')

plt.show()
```
